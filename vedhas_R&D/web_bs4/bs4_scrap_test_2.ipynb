```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
my_url="https://www.nintendo.com"
```

```python
Uclient = uReq(my_url)
page_html=Uclient.read()
Uclient.close()
page_soup = soup(page_html, "html.parser") ## lxml
```

```python
### all information on page
print(page_soup)
```

```python
## print in indentation (xml)
print(page_soup.prettify())
```
