```python
## import Packages
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
my_url="https://www.nintendo.com/games/game-guide" ## source as all games page 
```

```python
Uclient = uReq(my_url)
page_html=Uclient.read()
Uclient.close()
page_soup = soup(page_html, "html.parser") ## lxml
```

```python
#print(page_soup) # all information on page
print(page_soup.prettify()) # print in prety format 
```

```python
## print all contents in page with index
for index,child in enumerate(page_soup.descendants):
    print(index)
    print(child if child != '\n' else '')
```

##### sibling for each tag 

```python
## for body

for index,siblings in enumerate (page_soup.body.next_siblings):
    print(index)
    print(siblings.name if siblings != '\n' else '')
    print(siblings if siblings != '\n' else '')
```

```python
## for head 
for index,siblings in enumerate (page_soup.head.next_siblings):
    print(index)
    print(siblings.name if siblings != '\n' else '')
    print(siblings if siblings != '\n' else '')
```
